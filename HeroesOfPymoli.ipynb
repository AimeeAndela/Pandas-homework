## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
player_age = purchase_data.loc[:, ["SN", "Age"]]
player_age = player_age.drop_duplicates()
player_age
total_players = len(player_age)
total_players
```

```python
bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]
        
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

player_age["Age Summary"] = pd.cut(player_age["Age"], bins, labels=group_names)
player_age
```

```python
player_age = player_age.groupby("Age Summary")
total_age = player_age.count()
total_age
```

```python
avg_total_age = total_age["SN"]
avg_total_age
```

```python
percentage_age = round(total_age / total_players * 100, 2)
percentage_age
```

```python
percentage = percentage_age["SN"]
percentage
```

```python
age_summary_table = pd.DataFrame({"Total Count": [avg_total_age], 
                              "Percentage of Players": [percentage]})
age_summary_table
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_data["Age Summary"] = pd.cut(purchase_data["Age"], bins, labels=group_names)
purchase_data
```

```python
purchase_data_age = purchase_data.groupby("Age Summary")
count_age = purchase_data_age.count()
count_age
```

```python
purchase_count_age = count_age["Age"]
purchase_count_age
```

```python
avg_price_age = round(purchase_data_age["Price"].mean(), 2)
avg_price_age
```

```python
total_purchase_age = round(purchase_data_age["Price"].sum(), 2)
total_purchase_age
```

```python
avg_total_purchase_age = round(total_purchase_age / avg_total_age, 2)
avg_total_purchase_age
```

```python
total_purchase_summary_table = pd.DataFrame({"Purchase Count": [purchase_count_age], 
                              "Average Purchase Price": [avg_price_age], 
                                "Total Purchase Value": [total_purchase_age],
                                "Avg Total Purchase Per Person": [avg_total_purchase_age]})
total_purchase_summary_table
```
